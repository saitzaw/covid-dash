```
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```
virus_data = pd.read_csv('/content/drive/My Drive/Colab Notebooks/covid19_data.csv')
```

```
date = virus_data['Date']
daily_death = virus_data['daily death']
total_death = virus_data['total death']
```

```
plt.subplot(2,1,1)
plt.rcParams['figure.figsize'] = [20, 5]
time = np.arange(len(daily_death))
ax = plt.gca()
ax.set_xticks(time)
ax.xaxis.set_major_formatter(plt.NullFormatter())
plt.title("daily death in linear")
plt.plot(time, daily_death)
plt.grid()

plt.subplot(2,1,2)
plt.rcParams['figure.figsize'] = [20, 5]
time = np.arange(len(daily_death))
ax = plt.gca()
ax.set_xticks(time)
plt.title("daily death in log")
plt.plot(time, np.log(daily_death))
x_ticks = []
locs, labels = plt.xticks() 
new_xticks = date
plt.xticks(locs,new_xticks,rotation=45, horizontalalignment='right')
plt.grid()
```

```
daily_case = virus_data['daily case']
total_case = virus_data['total case']
```

```
plt.subplot(2,1,1)
plt.rcParams['figure.figsize'] = [18, 5]
time = np.arange(len(daily_case))
ax = plt.gca()
ax.set_xticks(time)
ax.xaxis.set_major_formatter(plt.NullFormatter())
plt.title("daily case in linear")
plt.plot(time, daily_case)
plt.grid()

plt.subplot(2,1,2)
plt.rcParams['figure.figsize'] = [18, 5]
time = np.arange(len(daily_case))
ax = plt.gca()
ax.set_xticks(time)
plt.title("dail case in log")
plt.plot(time, np.log(daily_case))
x_ticks = []
locs, labels = plt.xticks() 
new_xticks = date
plt.xticks(locs,new_xticks,rotation=45, horizontalalignment='right')
plt.grid()
```

```
#bar chart 
plt.rcParams['figure.figsize'] = [18,5]
time = np.arange(len(daily_death))
ax = plt.gca()
ax.set_xticks(time)
ax.xaxis.set_major_formatter(plt.NullFormatter())
plt.title("death toll")
plt.bar(time, daily_death,0.35,label="dail death")
plt.bar(time+0.35, total_death,0.35,label="total death")
x_ticks = []
locs, labels = plt.xticks() 
new_xticks = date
plt.xticks(locs,new_xticks,rotation=45, horizontalalignment='right')
plt.legend()
plt.grid()
```

```
#recovery 
report_data = pd.read_csv('/content/drive/My Drive/Colab Notebooks/covid19_report.csv')
total_summerize_labels = ['Total COVID-19 Infection Case', 'Total Deaths', 'Total Recovery']
active_case_labels = ['Active Infected Patient', 'Active Mild Condition', 'Active Critical']
pie_data_total = [
            np.average(report_data['total covid cases']),
            np.average(report_data['total deaths']),
            np.average(report_data['total recovery']),
            ] 
pie_data_active = [
            np.average(report_data['active infected patient']),
            np.average(report_data['active mild condition']),
            np.average(report_data['active critical'])
]
plt.subplot(1,2,1)
plt.pie(pie_data_total, labels=total_summerize_labels)    
plt.title("Infection, death and recovery")    
plt.subplot(1,2,2)
plt.pie(pie_data_active, labels=active_case_labels)    
plt.title("Infected patient, mild condition, critical") 

```
